# Email sender

### Prepare your email
- **Use HTML for create the body of your email**

```python
body_template = '''
<html>
<head>
    <style>
        /* Estilos para el cuerpo del correo */
        body {{
            font-family: Arial, sans-serif;
        }}
        .greetings {{
             margin-top: 10px;
             font-size: 11px;
        }}

        .paragraph {{
             margin-bottom: 5px;
             font-size: 11px;
        }}

        .warning {{
          font-size: 9px;
          margin-bottom: 20px;
        }}

        .contact {{
          margin-bottom: 3px;
          font-size: 9px;
          color: #808080;
        }}

    </style>
</head>
<body>
    <p class="greetings">Hi, {Receiver}!</p>

    <p class="paragraph">We hope this email finds you very well, and we appreciate your continued support.
    As part of our ongoing efforts to improve communication, we are conducting a test using Python for
    email delivery. This message may serve as a template for various contexts.
    Please review the following important information:</p>

    <p class="paragraph">This email is generated by a Python script to streamline communication processes.
    The script ensures efficiency and accuracy in delivering important updates,
    such as your affiliation credentials or whatever you need:</p>

    <p class="paragraph">This is an example:</p>

    <p class="parrafo">University: <strong>{university}</strong></br>Ranking university: <strong>{ranking}.</strong></p>

    <p class="parrafo">Look for more in: <strong>{link_uviversity}</strong></p>


    <p class="warning">*Hey: This is a message with other style, lot of fun doing it!</p>

    <p class="paragraph">While this system is designed to be robust, there are potential issues that may
    arise during mass email distribution. For instance, please verify that this email is sent to the
    correct address. Also, be aware of limitations, such as a maximum of 500 emails per day.
    Sending numerous emails immediately without introducing random delays may trigger spam
    filters or encounter server restrictions.</p>

    <p class="contact">If you are interested in my projects, <br/> send me an email <br/>Email-contact: luis.a7.guerrero@gmail.com</p>
</body>

</html>
'''
```

### Sending your email
- **Do not send this email for more than 500 emails**
- **Do not send this email without a random time sleep**

```python
# Set up SSL context for secure email connection
context = ssl.create_default_context()

# Connect to Gmail's SMTP server using SSL
with smtplib.SMTP_SSL('smtp.gmail.com', 465, context=context) as smtp:
    # Log in to the sender's email account
    smtp.login(sender, password)

    # Iterate through rows in the database DataFrame
    for index, row in dba_emails.iterrows():
        # Format the email body using a template
        mensaje = body_template.format(
            Receiver=row['Receiver'],
            university=row['university'],
            ranking=row['ranking'],
            link_uviversity=row['link_uviversity'],
        )

        # Create an EmailMessage object
        em = EmailMessage()
        em['From'] = from_name
        em['To'] = row['Email_receiver']
        em['Subject'] = subject_email
        em.add_alternative(mensaje, subtype='html')

        # Send the email, introducing random delays to avoid issues
        smtp.sendmail(from_name, row['Email_receiver'], em.as_string())
        time.sleep(random.uniform(5, 10))

# Display success message when all emails are sent
print("Email sent successfully!")
```
